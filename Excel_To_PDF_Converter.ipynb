```python
import win32com.client as win32
import os

def save_pdf(path_excel,output):
    print(f'Iniciando {path_excel}')
    
    file_name = os.path.splitext(os.path.basename(path_excel))[0]

    path_output_pdf = os.path.join(output,file_name +'.pdf')
    print(path_output_pdf)

    pdf_file = path_output_pdf

    # Inicializa o Excel
    excel = win32.Dispatch('Excel.Application')
    excel.Visible = False

    # Abre o arquivo Excel
    wb = excel.Workbooks.Open(path_excel)

    # Salva como PDF
    wb.ExportAsFixedFormat(0, pdf_file)

    # Fecha o arquivo e o Excel
    wb.Close()
    excel.Quit()

    print(f'Arquivo PDF gerado: {pdf_file}')

# Caminho do arquivo Excel

path_excel = r'F:\Vidya\Modec\MV31\Relatorio Tecnico\teste'
path_output = r'F:\Vidya\Modec\MV31\Relatorio Tecnico\prontos\module'

for files in os.listdir(path_excel):
    path_file = os.path.join(path_excel,files)
    print(path_file)
    save_pdf(path_file,path_output)

```
